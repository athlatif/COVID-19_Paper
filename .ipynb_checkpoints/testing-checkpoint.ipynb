```python
import numpy as np #importing numpy.
import pandas as pd #importing pandas.
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("testing_data.csv") 
df.drop(['cvd_death_rate', 'diabetes_prevalence', 'female_smokers','male_smokers','aged_70_older','aged_65_older','median_age'], axis=1)
# Preview the first 5 lines of the loaded data 
df.head()
```

```python
for col in df.columns: 
    print(col) 
```

```python
tests = []
cases = []
dates = []

firstcases = ['04-02-2020','18-01-2020']

# Germany

# first case 04-2-2020
Germany = df.loc[df['location'] == "Germany"]
Germany.head()


# US
US = df.loc[df['location'] == "United States"]
US.head()

# Saudi
Saudi = df.loc[df['location'] == "Saudi Arabia"]
Saudi.head()

# Korea
Korea = df.loc[df['location'] == "South Korea"]
Korea.head()

# Italy
Italy = df.loc[df['location'] == "Italy"]
Italy.head()

# Spain
Spain = df.loc[df['location'] == "Spain"]
Spain.head()
```

```python
germanytotaltest = []
germanytotalcases = []
germanytotaldates = []

firsttestingcondicted = []
firstcases = ['04-02-2020']

germanytotalcases = (list(Germany['total_cases_per_million']))
germanytotaltest = (list(Germany['new_tests_per_thousand']))
germanytotaldates = (list(Germany['date']))

#datetesting = (list(Germany['tests_units']))

cases.append(germanytotalcases[len(germanytotalcases)-1])
print(germanytotaltest)

test = []
date = []
for i in reversed(range(len(germanytotaltest))):
    if(np.isfinite(germanytotaltest[i])):
        test.append(germanytotaltest[i])
        
for i in range(len(germanytotaldates)):
    if(np.isfinite(germanytotaltest[i])):
        date.append(germanytotaldates[i])
        
tests.append(test[0])
dates.append(date[0])
```
