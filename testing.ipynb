```python
from scipy.interpolate import splrep, splev
from scipy.ndimage.filters import gaussian_filter1d

fig, ax1 = plt.subplots(figsize=(15, 8))

x =  np.linspace(0, len(dindexs)-1, num=len(dindexs))
color = 'tab:red'
ax1.set_xlabel('Days since the first case')
ax1.set_ylabel('Total Cases', color=color)
ax1.plot(mdays, dicases, color=color, label = "cases")
ax1.tick_params(axis='y', labelcolor=color)


ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ysmoothed2 = gaussian_filter1d(dindexs, sigma=2)
ax2.set_xlabel('Days since the first case')
ax2.set_ylabel('Index', color=color)
ax2.plot(mdays, ysmoothed2, color=color, label = "index")
ax2.tick_params(axis='y', labelcolor=color)

plt.suptitle('Saudi Arabia Index')

plt.legend()
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
