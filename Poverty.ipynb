```python
import numpy as np #importing numpy.
import pandas as pd #importing pandas.
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("occupation_data_UN.csv") 
# Preview the first 5 lines of the loaded data 
df.head()
df['Country or Area'].unique()
```

```python
Germany = df.loc[df['Country or Area'] == "Germany"]
Germany = Germany.loc[df['Sex'] == "Both Sexes"]
Germany = Germany.loc[df['Age'] == "Total"]

Occupationger = list(Germany['Occupation'])
Valueger = list(Germany['Value'])

Occupationger.pop(len(Valueger)-1)
Valueger.pop(len(Valueger)-1)

print(Occupationger)
print(Valueger)
```

```python
EssentialGer = []
RemoteGer = [] 

EssentialGer = Valueger[0] + Valueger[6] + Valueger[7] + Valueger[8] + Valueger[9] + Valueger[4]
RemoteGer = Valueger[1] + Valueger[2] + Valueger[3]  + Valueger[5]     
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt



objects = Occupationger
y_pos = np.arange(len(objects))
performance = Valueger

plt.barh(y_pos, np.log10(performance), align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('# number')
plt.title('Germany''s Common Occupations')
plt.xticks([0,1,2,3,4,5,6,7], ['1', '2.7k', '7.3k','20k', '55k', '148k','403k', '1M']) 

plt.show()
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt



objects = ['Essential Services','Remote-work Amenable Jobs']
y_pos = np.arange(len(objects))
performance = [EssentialGer,RemoteGer]
print(performance)
plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('# number')
plt.title('Remote-work Amenability of Occupations | Germany')


plt.show()
```
